Download Bert model ("dslim/bert-base-NER")

```python
#bert
tokenizer_bert = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
bert = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

ner_bert = pipeline("ner", model=bert, tokenizer=tokenizer_bert, device=device)
```

Download Roberta model ("xlm-roberta-large-finetuned-conll03-english")

```python
#roberta
tokenizer_roberta = AutoTokenizer.from_pretrained("xlm-roberta-large-finetuned-conll03-english")
roberta = AutoModelForTokenClassification.from_pretrained("xlm-roberta-large-finetuned-conll03-english")
ner_roberta = pipeline("ner", model=roberta, tokenizer=tokenizer_roberta, device=device)
```

```python
#split stories into phrases
splitted = []
for i in all_stories:
  splitted.append(i.split('\n'))
```

```python
#let's do NER (with roberta)
tot = []
for i in range(0, len(splitted)):
  res = ner_roberta(splitted[i])
  tot.append(res)
```

```python
#Function that split between selected labels and Misc
#for both return only the information needed (start, end, word, label), and reconstruct eventual split words
def prova_misc(lista, testo):
  etichette = {'I-PER': 'PER', 'I-ORG': 'ORG', 'I-LOC': 'LOC'}
  output = []
  remove = []
  for i in range(0, len(lista)):
    res = []
    char = 0
    misc = []
    for j in range(0, len(lista[i])):
      for k in range(0, len(lista[i][j])):
        entity = lista[i][j][k]['entity']
        if lista[i][j][k]['word'][0:1] != '▁':
          pass
        elif entity == 'I-MISC':
          word = lista[i][j][k]['word']
          start = lista[i][j][k]['start'] + char
          end = lista[i][j][k]['end'] + char
          label = 'I-MISC'
          try:
            while k+1 < len(lista[i][j]) and lista[i][j][k]['end'] == lista[i][j][k+1]['start']:
              word += lista[i][j][k+1]['word']
              end = lista[i][j][k+1]['end'] + char
              k += 1
          except:
            pass
          misc.append([start, end, word[1:], label])
        else:
          word = lista[i][j][k]['word']
          start = lista[i][j][k]['start'] + char
          end = lista[i][j][k]['end'] + char
          label = etichette[entity]
          try:
            while k+1 < len(lista[i][j]) and lista[i][j][k]['end'] == lista[i][j][k+1]['start']:
              word += lista[i][j][k+1]['word']
              end = lista[i][j][k+1]['end'] + char
              k += 1
          except:
            pass
          res.append([start, end, word[1:], label])
      char += len(testo[i][j])
    output.append(res)
    remove.append(misc)
  return output, remove
```

```python
#using the function upward to split between needed entity and Misc
out, misc = prova_misc(tot, splitted)
```

```python
#select only words, from the Misc results (create a list of words)
new = []
for i in misc:
  n = []
  for j in i:
    n.append(j[2])
  new.append(n)
```

```python
#run bert on the words classified Misc before
output_misc = []
for i in new:
  if i == []:
    output_misc.append([])
  else:
    bert_results = ner_bert(i)
    output_misc.append(bert_results)
```

```python
#put new labels on the Misc results
etichette = {'B-PER': 'PER', 'B-ORG': 'ORG', 'B-LOC': 'LOC', 'I-PER': 'PER', 'I-ORG': 'ORG', 'I-LOC': 'LOC'}

for i in range(0, len(output_misc)):
  for j in range(0, len(output_misc[i])):
    #print(output_misc[i][j])
    if output_misc[i][j] == []:
      pass
    else:
      if output_misc[i][j][0]['entity'][2:] == 'MISC':
        pass
      else:
        if output_misc[i][j][0]['score'] > 0.80: #set threshold
          misc[i][j][3] = etichette[output_misc[i][j][0]['entity']]
        else:
          pass
```

```python
#consider only the words classified PER, LOC or ORG by Bert (non consider words still classified MISC)
def delete_misc(lista):
  res = []
  for i in lista:
    r = []
    for j in i:
      if j[3][2:] == 'MISC':
        pass
      else:
        r.append(j)
    res.append(r)
  return res
```

```python
z = delete_misc(misc)
```

```python
#function to merge new Misc labelled right and order the list
def merge_misc(lista, misc):
  for i in range(0, len(misc)):
    if misc[i] == []:
      pass
    else:
      for j in misc[i]:
        lista[i].append(j)
    lista[i].sort()
  return lista
```

```python
out = merge_misc(out, z)
```

```python
#unify same entity splitted in two words (es: 'Jhon' 'Kennedy')
def try_unify(out):
  result = []
  for i in range(0, len(out)):
    o = []
    j = 0
    while j < len(out[i]):
      start = out[i][j][0]
      end = out[i][j][1]
      word = out[i][j][2]
      entity = out[i][j][3]
      while j+1 < len(out[i]):
        if (out[i][j][1] == out[i][j+1][0] or out[i][j][1]+1 == out[i][j+1][0]) and entity == out[i][j+1][3]:
          end = out[i][j+1][1]
          word += ' ' + out[i][j+1][2]
          j += 1
        else:
          break
      o.append([start, end, word, entity])
      j += 1
    result.append(o)
  return result
```

```python
merge = try_unify(out)
```

```python
#function to send request to BabelNet that return the ID correspondent
def add_BabelNet(word):
  i = 0
  r =requests.get('https://babelnet.io/v7/getSenses?lemma='+word+'&searchLang=EN&key=9e6c1b3e-c86d-4fe6-b72b-a5e059b9f169')
  r = r.json()
  try:
    while r[i]['properties']['pos'] != 'NOUN':
      if i < len(r) -1:
        i += 1
      else:
        i = len(r)
        break
  except:
    i = len(r)
  if i == len(r):
    return '---'
  else:
    return r[0]['properties']['synsetID']['id']
  
```

```python
#write all the information in a file
with open("./result_hw1.tsv", "w") as record_file: #open("/content/drive/MyDrive/EAI/result_hw1.tsv", "w") as record_file:
  for j in range(0, len(all_titles_short_stories)):
    dic = {}
    word = ''
    par = ''
    for i in merge[j]:
      if len(i[2]) == 0:
        pass
      else:
        par = i[2]
        s = par.split()
        word = ''
        for k in s:
          try:
            id = dic[k]
            word = k
          except:
            pass
        if word == '':
          id = add_BabelNet(i[2])
          dic[i[2]] = id
          word = i[2]
        #format: text_id, start_off, end_off, label, word_unique, id_BabelNet, word 
        record_file.write("%s\t%s\t%s\t%s\t%s\t%s\n" % (all_titles_short_stories[j], i[0], i[1], i[3], word, id))
```
