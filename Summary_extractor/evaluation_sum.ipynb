```python
path = "/content/drive/MyDrive/EAI/EAI2/Resume/"
```

```python
mine_path = ['The_Man_of_Adamant_mine.txt', 'The_She_bear_mine.txt', 'The_Voice_in_the_Night_mine.txt']
other_path = ['The_Man_of_Adamant_auto.txt', 'The_She_bear_auto.txt', 'The_Voice_in_the_Night_auto.txt']
#other_path = ['The_Man_of_Adamant_other.txt', 'The_She_bear_other.txt', 'The_She_bear_other2.txt', 'The_Voice_in_the_Night_other.txt']
```

```python
mine = open(path+mine_path[2],'r')
other = open(path+other_path[2],'r')
m = mine.readlines()
o = other.readlines()
```

```python
def merge_phrases(lista):
  l = lista[0]
  for i in range(1, len(lista)):
    l += ' ' + lista[i]
  return l
     
```

```python
string_m = merge_phrases(m)
string_o = merge_phrases(o)
```
