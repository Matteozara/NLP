```python
fairy_tales_path = "./FAIRY_TALE/texts/"
short_stories_path = "./SHORT_STORY/texts/"
```

```python
all_titles_fairy_tales = os.listdir(fairy_tales_path)
all_stories = []
for i in all_titles_fairy_tales:
  f = open(fairy_tales_path + i, "r")
  all_stories.append(f.read())

all_titles_short_stories = os.listdir(short_stories_path)
#short_stories = []
for i in all_titles_short_stories:
  f = open(short_stories_path + i, "r")
  all_stories.append(f.read())
```

```python
for i in all_titles_short_stories:
  all_titles_fairy_tales.append(i)

all_titles_short_stories = all_titles_fairy_tales
```

# Gensim

```python
from gensim.summarization import summarize
out = summarize(all_stories[0])
```

```python
res = []
for i in all_stories:
  out = summarize(i)
  res.append(out)
```

```python
with open("./result_hw2.tsv", "w") as record_file: #open("/content/drive/MyDrive/EAI/result_hw1.tsv", "w") as record_file:
  for i in range(0, len(res)):
    out = res[i].split('\n')
    phrases = ''
    for j in out:
      j = j.split(':')
      #print(j)
      if len(j[0]) < 4: #to discard phrases that aren't full (like only part of phrases)
        if len(phrases) == 0:
          phrases += j[0]
        else:
          phrases += ',' + j[0]
      #phrases += ',' + j[:2]
    record_file.write("%s\t%s\t\n" % (all_titles_short_stories[i], phrases))
  
```
